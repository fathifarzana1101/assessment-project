```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import pandas as pd
from datetime import datetime
```

```python
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
# Pretend to be a real Chrome browser
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get("https://www.amazon.in/s?k=laptops")
time.sleep(3)
```

```python
products = driver.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")
len(products)
```

```python
data = []
for p in products:
    # 1. Image URL
    try:
        image = p.find_element(By.TAG_NAME, "img").get_attribute("src")
    except:
        image = ""

    # 2. Title
    try:
        title = p.find_element(By.CSS_SELECTOR, "h2 a").text
    except:
        title = ""

    # 3. Rating
    try:
        rating = p.find_element(By.CSS_SELECTOR, "i span").text
    except:
        rating = ""

    # 4. Price
    try:
        price = p.find_element(By.CSS_SELECTOR, "span.a-price-whole").text
    except:
        price = ""

    # 5. Ad / Organic
    try:
        badge = p.find_element(By.CSS_SELECTOR, "span.s-label-popover-default").text
        if "Sponsored" in badge:
            result_type = "Ad"
        else:
            result_type = "Organic"
    except:
        result_type = "Organic"
    data.append([image, title, rating, price, result_type])
len(data)
```

```python
import pandas as pd
from datetime import datetime

# Convert list â†’ DataFrame
df = pd.DataFrame(data, columns=["Image URL", "Title", "Rating", "Price", "Ad/Organic"])

# Create timestamp
timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")

# Create file name
filename = f"amazon_laptops_{timestamp}.csv"

# Save CSV
df.to_csv(filename, index=False)

filename
```
